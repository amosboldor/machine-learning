```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Concerns housing values in suburbs of Boston.


## Attribute Information:

1. **CRIM:** per capita crime rate by town 
2. **ZN**: proportion of residential land zoned for lots over 25,000 sq.ft. 
3. **INDUS**: proportion of non-retail business acres per town 
4. **CHAS**: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise) 
5. **NOX**: nitric oxides concentration (parts per 10 million) 
6. **RM**: average number of rooms per dwelling 
7. **AGE**: proportion of owner-occupied units built prior to 1940 
8. **DIS**: weighted distances to five Boston employment centres 
9. **RAD**: index of accessibility to radial highways 
10. **TAX:** full-value property-tax rate per ``$10,000``
11. **PTRATIO:** pupil-teacher ratio by town 
12. **B:** 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town 
13. **LSTAT:** % lower status of the population 
14. **MEDV:** Median value of owner-occupied homes in $1000's

```python
data = pd.read_csv(os.getcwd() + '/boston_housing_data.csv',
                   names=["Del", "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSAT", "MEDV"])
del data['Del']
data
```

```python
average = data.MEDV.mean()
median = data.MEDV.median()
print("\nAverage home values across all Boston Suburbs in dollars: ${0:.2f}".format(average * 1000))
print("\nMedian home values across all Boston Suburbs in dollars: ${0:.2f}".format(median * 1000))
```

## The *median* and *average* home values across all Boston Suburbs in dollars.


**Average:** ``$22532.81``


**Median: ** ``$21200.00``

```python
def sift(x):
    if x["AGE"] < 50:
        return x["MEDV"]
    return None
```

```python
median = data.apply(sift, axis=1).median(skipna=None)
print("\nMedian home value of the suburb with the newest houses in dollars: ${0:.2f}".format(median * 1000))
```

## The *median* home value of the suburb with the newest houses in dollars.
**Median: ** ``$24400.00``

## The relationship between per-capita crime rate and the pupil-teacher ratio. Differentiated between whether or not the suburb is bounded by the Charles River.

**<span style="color:red">Red: </span>** Suburb bounded by the Charles River


**<span style="color:blue">Blue: </span>** Suburb not bounded by the Charles River

```python
plt.plot(data[data["CHAS"] == 0]["PTRATIO"], data[data["CHAS"] == 0]["CRIM"], 'ro')
plt.plot(data[data["CHAS"] == 1]["PTRATIO"], data[data["CHAS"] == 1]["CRIM"], 'bo')
plt.ylabel('per capita crime rate by town')
plt.xlabel('pupil-teacher ratio by town ')
plt.title('The relationship between per-capita crime rate and the pupil-teacher ratio.\nDifferentiated between whether or not the suburb is bounded by the Charles River.')
plt.show()
```

## The relationship between the proportion of black citizens and the distance to employment centers in Boston.

```python
plt.plot(data["B"], data["DIS"],'bo')
plt.ylabel('distance to employment centers in Boston')
plt.xlabel('proportion of black citizens')
plt.title('The relationship between the proportion of black citizens\n and the distance to employment centers in Boston.')
plt.show()
```

## The relationship between median value of owner-occuped homes and nitric oxide concentration
**<span style="color:green">Green: </span>** median home value and the proportion of non-retail business


**<span style="color:blue">Blue: </span>** median value of owner-occuped homes and nitric oxide concentration

```python
plt.plot(data["MEDV"], data["INDUS"], 'go', data["MEDV"],data["NOX"],'bo')
plt.ylabel('nitric oxides concentration\nand\nproportion of non-retail business')
plt.xlabel('median value of owner-occuped homes')
plt.title('The relationship between median value of owner-occuped homes and nitric oxide concentration\n along with median home value and the proportion of non-retail business.')
plt.show()
```
